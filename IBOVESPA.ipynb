```python
import pandas as pd

with open(r'/content/AcoesIndices_2023-12-27.csv', "r",encoding='latin1') as file:
    r_file = file.readlines()
    texto_completo = ''.join(r_file)
    linhas = texto_completo.split('\n')[2:]
    lista_de_listas = [linha.split(';') if ';' in linha else [linha] for linha in linhas]
    print(len(lista_de_listas))
```

```python


colunas = {'Empresa': [], 'Código': [], 'Índices dos quais participa': []}

# Preencher o dicionário com os dados
for linha in lista_de_listas:
  nova_linha = []
  for l, valor in enumerate(linha):

    if l == 0:
      colunas['Empresa'].append(valor)
    elif l == 1:
      colunas['Código'].append(valor)
    elif l>3:
      nova_linha.append(valor if l <1 else ' '.join(map(str, linha[3:])))
      colunas['Índices dos quais participa'].append(nova_linha)
del colunas['Código'][0]
del colunas['Empresa'][0]
del colunas["Empresa"][512]

for n in range(len(colunas["Índices dos quais participa"])):
  colunas["Índices dos quais participa"][n]=str(colunas["Índices dos quais participa"][n][0])





```

```python
df = pd.DataFrame(colunas)
pd.set_option('display.max_rows', None)
df
```

```python
indices=['IGCX', "ITAG", 'IGCT', 'IGNM']

empresas=[]

for indice in indices:
  for n in range(len(df['Índices dos quais participa'])):
    if indice in df['Índices dos quais participa'][n]:
      if df["Empresa"][n] not in empresas:
        empresas.append(df["Empresa"][n])
empresas

```
